```python
def calculate_rectangle_properties(width, height):
    """
    Calculates and prints the area and perimeter of a rectangle.

    Args:
        width (int or float): The width of the rectangle.
        height (int or float): The height of the rectangle.

    Returns:
        tuple: A tuple containing the area and perimeter of the rectangle.
    """
    area = width * height
    perimeter = 2 * (width + height)
    print(f"Area of Rectangle: {area}")
    print(f"Perimeter of Rectangle: {perimeter}")
    return area, perimeter

# Refactored calls using the new function
print("--- Rectangle 1 ---")
calculate_rectangle_properties(5, 10)

print("\n--- Rectangle 2 ---")
calculate_rectangle_properties(7, 12)

print("\n--- Rectangle 3 ---")
calculate_rectangle_properties(10, 15)
```

### Original (Legacy) Code and Performance Measurement

The original code uses nested loops to check if elements from `search_names` exist in `names`. This approach has a time complexity of O(n*m), where n is the length of `names` and m is the length of `search_names`. For large lists, this can become very inefficient.

```python
import time

names = [f"Name_{i}" for i in range(10000)] # Larger list for better performance demonstration
search_names = [f"Name_{i}" for i in range(5000, 15000, 100)] # Some present, some not

print("--- Original (Legacy) Code ---")
start_time = time.time()

for s in search_names:
    found = False
    for n in names:
        if s == n:
            found = True
            break # Found, no need to continue inner loop
    if found:
        print(f"{s} is in the list")
    else:
        print(f"{s} is not in the list")

end_time = time.time()
legacy_time = end_time - start_time
print(f"Original code execution time: {legacy_time:.6f} seconds")
```

### Refactored Code using Sets for Optimization

To significantly improve performance, we can convert the `names` list into a `set`. Set lookups (checking if an element is in the set) have an average time complexity of O(1). This reduces the overall complexity to O(n + m) (for creating the set and then iterating through `search_names`), which is much faster for larger datasets.

```python
import time

names = [f"Name_{i}" for i in range(10000)]
search_names = [f"Name_{i}" for i in range(5000, 15000, 100)]

print("\n--- Refactored Code with Sets ---")
start_time = time.time()

# Convert names to a set for O(1) average time complexity lookups
names_set = set(names)

for s in search_names:
    if s in names_set:
        print(f"{s} is in the list")
    else:
        print(f"{s} is not in the list")

end_time = time.time()
refactored_time = end_time - start_time
print(f"Refactored code execution time: {refactored_time:.6f} seconds")
```

### Performance Comparison

By comparing the execution times, you can observe the significant performance improvement achieved by using sets for element lookups. The set-based approach is considerably faster, especially as the size of the lists increases.

```python
def calculate_total(price):
    """
    Calculates the total price including an 18% tax.

    Args:
        price (int or float): The base price of the item.

    Returns:
        float: The total price after adding tax.
    """
    tax_rate = 0.18
    tax = price * tax_rate
    total = price + tax
    return total

# Refactored calls using the new function
print("--- Item 1 ---")
price1 = 250
total1 = calculate_total(price1)
print(f"Original Price: {price1}")
print(f"Total Price: {total1}")

print("\n--- Item 2 ---")
price2 = 500
total2 = calculate_total(price2)
print(f"Original Price: {price2}")
print(f"Total Price: {total2}")
```

### Task 4: Refactoring – Replacing Hardcoded Values with Constants

To improve maintainability, hardcoded 'magic numbers' should be replaced with named constants. This makes the code easier to understand and modify.

```python
# Original (Legacy) Code
print("--- Original (Legacy) Code ---")
print("Area of Circle:", 3.14159 * (7 ** 2))
print("Circumference of Circle:", 2 * 3.14159 * 7)

# Refactored Code with Constants
print("\n--- Refactored Code with Constants ---")
PI = 3.14159
RADIUS = 7

area_of_circle = PI * (RADIUS ** 2)
circumference_of_circle = 2 * PI * RADIUS

print(f"Area of Circle: {area_of_circle}")
print(f"Circumference of Circle: {circumference_of_circle}")
```
