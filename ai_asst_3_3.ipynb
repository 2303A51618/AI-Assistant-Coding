## Functions for Charges

### Subtask:
Create Python functions to encapsulate the logic for calculating `energy_charge` and introduce a new `fixed_charge` calculation.

```python
# Define fixed rates for different customer types
residential_fixed_charge = 10.0
commercial_fixed_charge = 50.0

def calculate_energy_charge(units_consumed, customer_type, residential_rate, commercial_rate):
    """
    Calculates the energy charge based on units consumed and customer type.
    """
    energy_charge = 0.0
    if customer_type == 'Residential':
        energy_charge = units_consumed * residential_rate
    elif customer_type == 'Commercial':
        energy_charge = units_consumed * commercial_rate
    else:
        print(f"Warning: Unknown customer type '{customer_type}'. Energy charge set to 0.")
    return energy_charge

def calculate_fixed_charge(customer_type, residential_fixed_charge, commercial_fixed_charge):
    """
    Calculates the fixed charge based on the customer type.
    """
    fixed_charge = 0.0
    if customer_type == 'Residential':
        fixed_charge = residential_fixed_charge
    elif customer_type == 'Commercial':
        fixed_charge = commercial_fixed_charge
    else:
        print(f"Warning: Unknown customer type '{customer_type}'. Fixed charge set to 0.")
    return fixed_charge

# Example usage with existing values
example_units_consumed = units_consumed
example_customer_type = customer_type

calculated_energy_charge = calculate_energy_charge(example_units_consumed, example_customer_type, residential_rate, commercial_rate)
calculated_fixed_charge = calculate_fixed_charge(example_customer_type, residential_fixed_charge, commercial_fixed_charge)

print(f"\n--- Charges Calculated with Functions ---")
print(f"Units Consumed: {example_units_consumed}")
print(f"Customer Type: {example_customer_type}")
print(f"Calculated Energy Charge: {calculated_energy_charge:.2f}")
print(f"Calculated Fixed Charge: {calculated_fixed_charge:.2f}")
total_bill = calculated_energy_charge + calculated_fixed_charge
print(f"Total Bill (Energy + Fixed): {total_bill:.2f}")
```

**Reasoning**:
The subtask requires calculating the energy charge based on customer type and units consumed, using conditional statements and including error handling for unknown customer types. This code block will implement that logic.



```python
energy_charge = 0.0

if customer_type == 'Residential':
    energy_charge = units_consumed * residential_rate
elif customer_type == 'Commercial':
    energy_charge = units_consumed * commercial_rate
else:
    print(f"Unknown customer type: {customer_type}. Energy charge set to 0.")
    energy_charge = 0.0

print(f"Calculated Energy Charge: {energy_charge}")
```
