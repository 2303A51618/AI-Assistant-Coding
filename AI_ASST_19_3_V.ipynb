## C++ Equivalent of the Student Class

This C++ code defines a `Student` class with `name` (string) and `age` (integer) as private member variables. It includes:

-   **Constructor:** Initializes the `name` and `age` of a `Student` object.
-   **`display()` method:** Prints the student's name and age.
-   **`main()` function:** Demonstrates how to create `Student` objects and call their `display()` method.

To compile and run this code, you would typically save it as a `.cpp` file (e.g., `student.cpp`) and use a C++ compiler like g++:

```bash
g++ student.cpp -o student
./student
```


````python
%%writefile student.cpp
#include <iostream>
#include <string>

class Student {
private:
    std::string name;
    int age;

public:
    // Constructor
    Student(std::string studentName, int studentAge) {
        name = studentName;
        age = studentAge;
    }

    // Method to display student information
    void display() {
        std::cout << "Student Name: " << name << std::endl;
        std::cout << "Student Age: " << age << std::endl;
    }
};

int main() {
    // Create Student objects
    Student student1("Alice", 20);
    Student student2("Bob", 22);

    // Call the display method for each student
    std::cout << "Displaying Student 1 Info:" << std::endl;
    student1.display();
    std::cout << "\nDisplaying Student 2 Info:" << std::endl;
    student2.display();

    return 0;
}
````

### Python Equivalent of `isPrime()` Method

This Python function `is_prime(n)` checks if a given integer `n` is a prime number. It follows the same logic as the provided Java method:

-   Numbers less than or equal to 1 are not prime.
-   It iterates from 2 up to the square root of `n`. If any number in this range divides `n` evenly, `n` is not prime.
-   If no such divisor is found, `n` is prime.

```python
def is_prime(n):
    if n <= 1:
        return False
    # Loop from 2 up to the square root of n
    # We only need to check divisibility up to sqrt(n)
    # because if n has a divisor greater than its square root,
    # it must also have a divisor smaller than its square root.
    i = 2
    while i * i <= n:
        if n % i == 0:
            return False
        i += 1
    return True
```

## Python Implementation of Bubble Sort

This Python function `bubble_sort(arr)` sorts a list of numbers in ascending order using the bubble sort algorithm. The logic directly translates from the pseudocode provided:

-   **Outer loop (`i`):** Controls the number of passes through the list. With each pass, the largest unsorted element "bubbles up" to its correct position at the end of the unsorted portion.
-   **Inner loop (`j`):** Compares adjacent elements and swaps them if they are in the wrong order. The range of this loop decreases with each outer pass because elements at the end are already sorted.

```python
def bubble_sort(arr):
    n = len(arr)
    # Traverse through all array elements
    for i in range(n):
        # Last i elements are already in place
        for j in range(0, n-i-1):
            # Traverse the array from 0 to n-i-1
            # Swap if the element found is greater
            # than the next element
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
    return arr
```

### Test Case

The following code block defines a sample list and then applies the `bubble_sort()` function to sort it, displaying the original and sorted lists.


````python
def bubble_sort(arr):
    n = len(arr)
    # Traverse through all array elements
    for i in range(n):
        # Last i elements are already in place
        for j in range(0, n-i-1):
            # Traverse the array from 0 to n-i-1
            # Swap if the element found is greater
            # than the next element
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
    return arr

# Test with a sample list
sample_list = [64, 34, 25, 12, 22, 11, 90]
print(f"Original list: {sample_list}")

sorted_list = bubble_sort(sample_list)
print(f"Sorted list: {sorted_list}")

# Another test case
sample_list_2 = [5, 1, 4, 2, 8]
print(f"Original list 2: {sample_list_2}")
sorted_list_2 = bubble_sort(sample_list_2)
print(f"Sorted list 2: {sorted_list_2}")

# Edge case: already sorted
sample_list_3 = [1, 2, 3, 4, 5]
print(f"Original list 3 (already sorted): {sample_list_3}")
sorted_list_3 = bubble_sort(sample_list_3)
print(f"Sorted list 3: {sorted_list_3}")

# Edge case: reverse sorted
sample_list_4 = [5, 4, 3, 2, 1]
print(f"Original list 4 (reverse sorted): {sample_list_4}")
sorted_list_4 = bubble_sort(sample_list_4)
print(f"Sorted list 4: {sorted_list_4}")
````

## SQL to Pandas Conversion: Filtering and Selecting

This section demonstrates how to convert the SQL query `SELECT name, salary FROM employees WHERE salary > 50000;` into a Pandas DataFrame operation.

First, a sample DataFrame named `employees_df` is created to simulate an 'employees' table with 'name', 'age', and 'salary' columns.

Then, the Pandas operation filters the `employees_df` to include only employees with a 'salary' greater than 50000 and selects the 'name' and 'salary' columns, mimicking the SQL query's behavior.


````python
import pandas as pd

# Create a sample DataFrame resembling an 'employees' table
employees_df = pd.DataFrame({
    'name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'age': [30, 24, 35, 40, 28],
    'salary': [60000, 45000, 75000, 50000, 62000]
})

print("Original DataFrame:")
display(employees_df)

# Convert the SQL query 'SELECT name, salary FROM employees WHERE salary > 50000;' to Pandas
# 1. Filter rows where 'salary' > 50000
# 2. Select 'name' and 'salary' columns
result_df = employees_df[employees_df['salary'] > 50000][['name', 'salary']]

print("\nResulting DataFrame (salary > 50000):")
display(result_df)
````

## C Equivalent of Linear Search

This C code implements the `linearSearch` function, which searches for a `target` element within an integer array `arr` of a given `size`. It iterates through each element of the array, comparing it with the target. If the target is found, it returns the index of the first occurrence; otherwise, it returns `-1`.

Key differences from Python include:
-   **Data Types:** Explicit type declarations (`int`, `int*` for array pointers). C requires you to specify the type of each variable.
-   **Array Handling:** Arrays are passed as a pointer to their first element (`int arr[]` or `int* arr`) along with their `size`. C does not automatically track array length.
-   **Memory Management:** Arrays are typically allocated on the stack (for fixed-size arrays) or heap (for dynamic arrays using `malloc`/`free`). In this example, `arr` in `main` is a stack-allocated array. Python handles memory automatically.
-   **Function Signature:** The function signature includes explicit types for all parameters and the return value.
-   **Libraries:** Uses `<stdio.h>` for input/output functions like `printf`.

    }
